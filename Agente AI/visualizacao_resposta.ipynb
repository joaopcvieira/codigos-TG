```python
import pandas as pd
import numpy as np

# Le do arquivo txt a matriz A
# Usar delimitador de um único caractere. Espaços após a vírgula são ignorados automaticamente.
A = np.loadtxt('matriz_dematel.txt', delimiter=',')

A



df_fatores = pd.read_csv('fatores.csv')
fatores = df_fatores['fator'].tolist()

# Verifica consistência entre número de fatores e dimensão da matriz A
if A.shape[0] != len(fatores):
    raise ValueError(f'Tamanho da matriz A ({A.shape[0]}) difere do número de fatores ({len(fatores)}).')

# Cria o DataFrame rotulado diretamente a partir de A
df_esqueleto = pd.DataFrame(A, index=fatores, columns=fatores)
df_esqueleto
```
